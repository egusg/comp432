# **Mean Squared Error**

The *Mean Squared Error* (MSE) is an objective function often used in regression problems. It measures "how far" is the prediction of the machine learning model from the targeted one. It is defined as:

$
MSE = \frac{1}{N} \sum_{n=1}^N (y_n - f(x_n) )^2
$

where $Y = [y_1, y_2, ..., y_n, ..., y_N]$ contains the **targets** (labels), $X = [x_1, x_2, ..., x_n, ..., x_N]$ the **inputs**, and $f(x)$ is the input-output mapping **function** learned by the machine leaning algorithm.



```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mse(y, pred):
  return ((y - pred)**2).mean()
```

Let's now draw the targets points from a sine function:

```python
def get_dataset(N):
  "This function creates a dataset wih N examples."
  x = np.random.uniform(low=0.0, high=2 * math.pi, size=N)
  return x, np.sin(x)
```

```python
# Let's draw the training and test set
x_train, y_train = get_dataset(25)
```

```python
# We can plot the training samples
plt.plot(x_train, y_train, 'o', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['targets (labels)'])
plt.show()
```

Let's now assume that the machine learning algorithm has learned a slightly different function:

$f(x) = 0.8 sin(x)$

The mean squared error can be computed as follows:

```python
predictions = 0.8 * np.sin(x_train)
mse_value = mse(y_train, predictions)
print(mse_value)
```

```python
# We can plot the training samples
plt.plot(x_train, y_train, 'o', color='blue')
plt.plot(x_train, predictions, 'o', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['targets (labels)', 'predictions'])
plt.show()
```

Let's try now assume that my machine learning algorithms is really bad and it found a function very far away from the targets

$f(x) = log(x)$

```python
predictions = 0.8 * np.log(x_train)
mse_value = mse(y_train, predictions)
print(mse_value)
```

```python
# We can plot the training samples
plt.plot(x_train, y_train, 'o', color='blue')
plt.plot(x_train, predictions, 'o', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['targets (labels)', 'predictions'])
plt.show()
```

As you can see, the MSE is much higher in this case. If the targets and the predictions are the same, the MSE is zero:

```python
predictions = np.sin(x_train)
mse_value = mse(y_train, predictions)
print(mse_value)
```

**Summary**: MSE is a simple objective function that can tell us how good a solution is in a regression problem.
