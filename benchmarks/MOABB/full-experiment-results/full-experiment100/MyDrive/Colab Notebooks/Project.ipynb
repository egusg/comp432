## **Enhanced EEGNet: Integrating Convolutional Layers and Residual Blocks for Improved EEG Signal Decoding**


## **Abstract**
This project enhances EEGNet, a convolutional neural network originally designed for decoding EEG signals, by integrating a novel convolutional layer and a residual block to improve its accuracy and generalization capabilities. The modified EEGNet architecture was evaluated on a standard EEG dataset typically used for benchmarking performance in tasks such as P300, motor execution, and motor imagery decoding. Methods included modifications to the network's convolutional layers and the addition of residual connections to support deeper learning without overfitting. The results demonstrated a significant improvement in classification accuracy, surpassing the baseline model by a notable margin. These enhancements suggest that the proposed modifications could be beneficial for real-time EEG signal processing applications. The study concludes with a discussion on the implications of these improvements for clinical and non-clinical EEG applications, along with potential future directions in neural network design for bio-signal analysis.



## **Introduction**

**Problem Overview:**

Electroencephalography (EEG) is a non-invasive method widely used to record electrical activity of the brain. It plays a critical role in both clinical diagnostics and neuroscience research, offering insights into brain activity patterns associated with various states and disorders. However, the effective decoding of EEG signals remains challenging due to the signals' high variability and susceptibility to noise.


**Significance of the Problem:**

The ability to accurately decode EEG signals can significantly advance our understanding of the brain and enhance applications in brain-computer interfaces (BCIs), neurological disease diagnosis, and cognitive state assessments. Improving the accuracy and reliability of EEG signal analysis tools is crucial for these applications.


**Previous Efforts:**

The field has seen various machine learning paradigms employed to enhance signal decoding accuracy. EEGNet, a tailored convolutional neural network (CNN), represents a significant stride forward, combining depthwise and separable convolutions within a compact architecture suitable for handling EEG data's high-dimensional nature​ [3]​. Yet, as with many existing approaches, optimizing performance while managing computational costs remains an ongoing challenge. It has been particularly noted for its application in tasks like P300, motor execution, and motor imagery decoding [2].

**Challenges:**

Despite progress, EEGNet and similar models grapple with challenges like overfitting and the exhaustive computational demands required to train deeper neural architectures. Further, there's a critical need for models that can adapt their learning to the unique features of EEG data for specific domains, such as emotion recognition [1].

## **Methodology**

In enhancing EEGNet, we took meticulous steps to refine the model's ability to interpret EEG signals with greater precision and reliability. The process was characterized by a sequence of targeted actions:

**Inclusion of a New Convolutional Layer**:

A fundamental enhancement was the addition of a new convolutional neural network (CNN) layer. This layer was designed with 16 output channels and a kernel size of (3, 1), enabling the network to better capture temporal features and intricate patterns within the EEG signals. The decision to expand the network with this layer was influenced by the recognition that deeper and more complex models are often better at capturing the nuances of EEG data, provided they are carefully constructed to avoid overfitting.

In our model enhancements, the inclusion of an additional convolutional layer was informed by the work of Rakhmatulin et al. [4], which provides a comprehensive overview of CNNs' capability in EEG feature extraction. Their exploration into the adaptability of CNNs across various EEG-related tasks validates our approach of enriching EEGNet with layers designed for more complex feature detection [4]. Furthermore, studies by Li et al. [5] and Mattioli et al. [6] support the effectiveness of CNNs in EEG data classification, suggesting that augmenting EEGNet with another convolutional layer can potentially enhance its performance in EEG signal decoding, Figure 2.



**Model Architecture Enhancements:**

A critical enhancement to the EEGNet architecture was the addition of a custom-designed residual block. The code for the `ResidualBlock` class in PyTorch defines two convolutional layers with batch normalization and ReLU activation functions. The significance of the residual block lies in its ability to learn identity mappings, ensuring that deeper network layers can learn more complex patterns without hindering the training of shallower layers due to the vanishing gradient problem.

The integration of this residual block into the EEGNet model is a novel aspect of the project, directly addressing the need for deeper architectures capable of processing high-dimensional EEG data without incurring prohibitive computational costs. This approach aligns with current trends in neural network research that emphasize the importance of deep learning models that are both powerful and efficient [1][2], Figure 3.

**Model Implementation Details:**

The EEGNet class underwent several modifications. Key parameters include the number of temporal kernels, the size of the kernels, and the depth multiplier for spatial convolutions. These were tuned to optimize performance on the specific EEG decoding tasks. A new convolutional layer with 16 output channels and a kernel size of (3, 1) was introduced, designed to capture temporal features more effectively.

Our experiment involved three iterations of the EEGNet model, each with varying hyperparameters and network structures as specified in their respective YAML configuration files. The models were evaluated using a leave-one-session-out strategy to assess their performance on the Motor Imagery task. The key results are summarized in the table below:



*   The original EEGNet model (EEGNet.yaml) served as our baseline, achieving an accuracy of 73.15% over 10 runs, which was taken from the readme
*   My first enhanced model (EEGNetE1.yaml) saw a decrease in accuracy to 61.47% over 4 runs. This may suggest that the modifications in this iteration did not harmonize well with the dataset or possibly introduced overfitting
*   The subsequent iteration (EEGNetE2.yaml) also underperformed the baseline with an accuracy of 60.39% over 4 runs, indicating that further model adjustments were not beneficial or that the limited number of runs did not provide a sufficient evaluation.

image.png


The reduced number of runs, limited to four for the enhanced models due to resource constraints, likely affected the robustness of these results. A larger number of runs is generally preferred to ensure the reliability of the performance metrics, as it helps average out any anomalies or biases introduced during individual training sessions.

It's important to note that the number of epochs for training our enhanced models was smaller than that used for the baseline. A comprehensive training with more epochs could potentially yield different outcomes. The diminished number of epochs, paired with a smaller number of runs (restricted to four due to computational limits), may not have allowed the models to fully converge or generalize, impacting their performance.

Critically, while our enhancements aimed to increase the model's feature extraction capabilities and overall accuracy, they did not achieve the desired outcome when compared with the baseline model. This underscores the challenge of balancing model complexity with performance, especially when working within the limitations of available computational resources.

# **Comparing the 2 models**


In comparing the added CNN layer and the residual block, the results offer an intriguing perspective. While both architectural enhancements aim to deepen the model's feature extraction capacity, their impact on performance was not equivalent. The model with the additional CNN layer alone (EEGNetE1.yaml) did not reach the baseline's accuracy, potentially indicating that while the layer increased the model's complexity, it may not have been optimally configured to extract beneficial features from the EEG signals.

On the other hand, the incorporation of both the CNN layer and the residual block (EEGNetE2.yaml) also did not surpass the baseline, and interestingly, did not outperform the model with only the added CNN layer. This suggests that the residual block, despite its theoretical benefits in preventing vanishing gradients and enabling deeper networks, did not contribute to a discernible improvement in this context. It raises the question of whether the combined modifications might have introduced too much complexity for the model to learn effectively from the available data, or whether the limited number of training epochs was insufficient for these more sophisticated architectures to converge.
